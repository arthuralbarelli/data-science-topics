# Libraries and modules

```python
import yaml
import pandas as pd

from sklearn.model_selection import train_test_split

from pathlib import Path
```

# Opening YAML file

```python
config_path = "../script/config.yml"

try:
    with open(config_path, 'r') as f:
        config = yaml.safe_load(f)
except Exception as e:
    print("Cannot open config file.")
```

```python
print(config)
```

# Acessing configuration parameters

```python
raw_data_file = config["etl"]["raw_data_file"]
processed_path = Path(config["etl"]["processed_path"])
test_size = config["etl"]["test_size"]
train_file = config["train"]["train_file"]
test_file = config["predict"]["test_file"]
random_state = config["etl"]["random_state"]
```

# Testing usage

```python
df_recruitment = pd.read_csv(raw_data_file)
```

```python
train, test = train_test_split(df_recruitment, test_size=test_size,
                               random_state=random_state)

train.to_csv(train_file, index=False)
test.to_csv(test_file, index=False)
```
